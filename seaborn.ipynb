# Correlation with Heatmap 

A correlation heatmap uses colored cells, typically in a monochromatic scale, to shaow a 2D correlation matrix(table) between two discrete dimensions or event types. It is very important feature selection.

```python
numeric_df = df.select_dtypes(include=['number'])   # Select only numeric columns
correlation_matrix = numeric_df.corr()   # Calculate correlation matrix
```

```python
sns.heatmap(correlation_matrix)
```

# Joinplot 

A joinplot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

# Univariate Analysis

```python
sns.jointplot(x='tip', y='total_bill',data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='size',data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='size',data=df, kind='reg')
```

# Pairplot 

A "pair plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this , showing all variables paired with all other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

```python
df['time'].value_counts()
```

# Distplot
Dist plot helps us to check the distribution of columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.displot(df['tip'])
```

```python
sns.histplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

#  Categorical Plots

Seaborn also helps us in doing the analysis on categorical data points. In this section we will discuss about.
- boxplot
- violinplot
- countplot
- barplot

```python
## Countplot
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
## Bar plot
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
df.head(10)
```

```python
sns.barplot(y='tip',x='sex',data=df)
```

# Boxplot
A box and whisker plot(sometimes called a boxplot) is a graph that presents information from a five- number summary.

```python
sns.boxplot(x='time',y='total_bill', data=df, palette='rainbow',hue='time')
```

```python
sns.boxplot(x='day',y='total_bill', data=df,hue='sex')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# Categorize my data based on some other categories
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

# Violin plot

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot.

```python
sns.violinplot(x="total_bill", y="day", data=df, palette='rainbow', hue='day')
```
