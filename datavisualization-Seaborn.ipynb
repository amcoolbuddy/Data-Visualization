# Line Plot Method

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


var = [1,2,3,4,5,6,7]
var_1=[2,3,1,4,5,6,7]
x_1 = pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var", y="var_1",data=x_1)
plt.show()
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins").head(30)
y_1
```

```python
sns.lineplot(x="bill_length_mm", y="flipper_length_mm", data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=y_1, hue="sex", size=20, style="sex",palette="rainbow",markers=["o",">"])
plt.grid()
plt.title("Lineplot")
plt.show()
```

# Barplot

```python
y_1 = sns.load_dataset("penguins").head(100)
y_1
```

```python
sns.barplot(x=y_1.island, y=y_1.bill_length_mm, palette="rainbow",data=y_1,hue='sex')
plt.show()
```

```python
sns.set(style="darkgrid")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=y_1.island, y=y_1.bill_length_mm, palette="dark",data=y_1,hue='sex',order=order_1, hue_order=["Female","Male"], errorbar=('ci',100)
           ,saturation=1,err_kws={'color':'g','linewidth':'5'},alpha=0.8)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=y_1.bill_depth_mm, y=y_1.bill_length_mm,data=y_1,hue='sex', orient='v')
plt.show()
```

# Violin Plot

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more)
categorical variables such that those distributions can be compared.



```python
y_1 = sns.load_dataset("tips")
y_1
```

```python
sns.violinplot(data=y_1, x="sex", y= "total_bill", palette="rainbow", hue='day')
plt.show()
```

```python
sns.violinplot(data=y_1, x="time", y= "total_bill", palette="rainbow", order=["Dinner","Lunch"],hue='sex',)
plt.show()
# inner="quart" still to understand
```

```python
sns.violinplot(data=y_1, x="day", y= "total_bill", palette="rainbow",hue='sex', split= True)
plt.show()
```

```python
sns.violinplot(data=y_1, x="time", y= "total_bill", palette="rainbow",hue='sex', split= True)
plt.show()
```

```python
sns.violinplot(data=y_1, x="total_bill", palette="rainbow", hue='sex')
plt.show()
```

# Box Plot

```python
y_1 = sns.load_dataset("tips")
y_1
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=y_1, y="tip", x="day", hue="smoker", palette="Paired", order=["Sun","Sat","Fri","Thur"], showmeans=True,
            meanprops={"marker":"o","markeredgecolor":"b"}, linewidth=1)
plt.show()
```

# Factor Plot

```python
y_1 = sns.load_dataset("tips").head(50)
y_1
```

```python
sns.catplot(x="tip",y="size",data=y_1, hue='sex',palette="dark:green",kind="bar")
plt.show()
```

# Styling Plot

```python
y_1 = sns.load_dataset("tips")
y_1
```

```python
sns.set_style("white")
plt.figure(figsize=(12,3))
sns.barplot(data=y_1,x="day",y="tip", palette="dark", hue='sex')
sns.despine()
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(3,10))
sns.barplot(data=y_1,x="day",y="tip", palette="dark", hue='sex')
sns.despine()
plt.show()
```

```python
sns.set_style("white")
sns.set_context("poster",font_scale=2)
sns.barplot(data=y_1,x="day",y="tip",palette='cool')
sns.despine()
plt.show()
```
